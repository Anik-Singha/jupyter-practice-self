## For Loop

```python
integers = [1,2,3,4,5]
```

```python
for number in integers:
    print(number)
```

```python
for number in integers:
    print('*'*number)
```

```python
for number in integers:
    print('*'*(len(integers)-number+1))
```

```python
for number in integers:
    print(('*'*number)+' '+ ('*'*(len(integers)-number+1)))
```

```python
for number in integers:
    spaces = len(integers) - number
    stars = number
    print(' ' * spaces + '*' * stars)
```

```python
n = len(integers)

# Find largest odd number <= n
max_odd = n if n % 2 != 0 else n - 1

rows = (max_odd + 1) // 2

for i in range(rows):
    stars = 2 * i + 1
    spaces = (max_odd - stars) // 2
    print(" " * spaces + "*" * stars)
```

```python
n = 7
max_odd = n if n % 2 != 0 else n-1
rows = (max_odd +1 ) //2

for i in range(rows):
    stars = 2 * i + 1
    spaces = rows - (i + 1)
    print(" " * spaces + "*" * stars)
```

## List Values

```python
ice_cream_dict = {'name': 'Anik Singha', 'price': 200, 'weight': 200 ,'fav_ice_cream': ['Chocolate','Vanilla']}
```

```python
for cream in ice_cream_dict.keys():
    print(cream)
```

```python
for cream in ice_cream_dict.keys():
    print(ice_cream_dict[cream])
```

```python
for cream in ice_cream_dict.values():
    print(cream)
```

## Nested For Loop

```python
val1 = [1,2,3,4,5]
val2 = [4,5,6,7,8]
```

```python
count = 1
for num1 in val1:
    for num2 in val2:
        print(count,' || ', num1, '*', num2 ,' = ', num1*num2)
        count += 1
```
