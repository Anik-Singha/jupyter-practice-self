## For Loop

```python
integers = [1,2,3,4,5]
```

```python
for number in integers:
    print(number)
```

```python
for number in integers:
    print('*'*number)
```

```python
for number in integers:
    print('*'*(len(integers)-number+1))
```

```python
for number in integers:
    print(('*'*number)+' '+ ('*'*(len(integers)-number+1)))
```

```python
for number in integers:
    spaces = len(integers)-number + 1
    stars = number
    print(' ' * spaces + '*' * stars)
```

```python
for number in integers:
    print(' ' * (len(integers) - number - 1),end='')
    print('*' * (number))
```
