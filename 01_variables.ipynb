Inserting Multiple Variables at a time

```python
x,y,z = 'Chocolate', 'Mint', 'Cocoa'
print(x,y,z)
```

```python
x = y = z = 'Root Leaf Branch'
print(x,y,z, sep = '\n')
```

```python
ice_cream = ['Chocolate', 'Mint', 'Cocoa']

x,y,z = ice_cream

print(x,y,z, sep = '\n')
```
